```python
import pandas as pd

# 로그 데이터
q_log_lines = """
layer grad norm: 1177.0 layer weight norm: 767.0
layer grad norm: 1261.0 layer weight norm: 767.0
layer grad norm: 1158.0 layer weight norm: 767.0
layer grad norm: 1028.0 layer weight norm: 766.5
layer grad norm: 654.5  layer weight norm: 767.0
layer grad norm: 1163.0 layer weight norm: 766.5
layer grad norm: 1185.0 layer weight norm: 767.0
layer grad norm: 1570.0 layer weight norm: 764.0
layer grad norm: 768.5  layer weight norm: 766.5
layer grad norm: 799.0  layer weight norm: 767.0
layer grad norm: 1703.0 layer weight norm: 767.0
layer grad norm: 632.5  layer weight norm: 766.5
""".strip().split("\n")

v_log_lines = """
layer grad norm: 1287.0 layer weight norm: 762.0
layer grad norm: 788.0  layer weight norm: 763.0
layer grad norm: 585.5  layer weight norm: 763.5
layer grad norm: 946.0  layer weight norm: 764.0
layer grad norm: 1213.0 layer weight norm: 764.0
layer grad norm: 842.0  layer weight norm: 763.5
layer grad norm: 1196.0 layer weight norm: 764.0
layer grad norm: 1570.0 layer weight norm: 764.0
layer grad norm: 1538.0 layer weight norm: 764.5
layer grad norm: 1447.0 layer weight norm: 764.5
layer grad norm: 925.5  layer weight norm: 765.0
layer grad norm: 1032.0 layer weight norm: 765.5
""".strip().split("\n")

# 데이터를 저장할 리스트
q_grad_norms, q_weight_norms = [], []
v_grad_norms, v_weight_norms = [], []

# q_log_lines 처리
for line in q_log_lines:
    parts = line.split("layer grad norm: ")[1].split(" layer weight norm: ")
    q_grad_norms.append(float(parts[0]))
    q_weight_norms.append(float(parts[1]))

# v_log_lines 처리
for line in v_log_lines:
    parts = line.split("layer grad norm: ")[1].split(" layer weight norm: ")
    v_grad_norms.append(float(parts[0]))
    v_weight_norms.append(float(parts[1]))

# 데이터프레임 생성
df_q = pd.DataFrame({'Layer': list(range(len(q_grad_norms))), 'Q_Grad_Norm': q_grad_norms, 'Q_Weight_Norm': q_weight_norms})
df_v = pd.DataFrame({'Layer': list(range(len(v_grad_norms))), 'V_Grad_Norm': v_grad_norms, 'V_Weight_Norm': v_weight_norms})

# 데이터 출력
print("Q Projection Grad & Weight Norms")
print(df_q.to_string(index=False))  # to_string 사용하여 모든 행 출력
print("\nV Projection Grad & Weight Norms")
print(df_v.to_string(index=False))  # to_string 사용하여 모든 행 출력
```
